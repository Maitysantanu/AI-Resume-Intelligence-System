```python
with open("data/job_required_skills.json") as f:
    role_data=json.load(f)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,1))

job_vectors = vectorizer.fit_transform(df["skills"])
```

```python
model = SentenceTransformer("all-mpnet-base-v2")

job_embeddings = model.encode(df["skills"].tolist())
```

```python
with open("model/tfidf_vectorizer.pkl","wb") as f:
    pickle.dump(vectorizer,f)

with open("model/job_vectors.pkl","wb") as f:
    pickle.dump(job_vectors,f)

np.save("model/job_embeddings.npy",job_embeddings)

print("Models saved successfully")
```
